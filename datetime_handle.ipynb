```python
import numpy as np
import pandas as pd
```

```python
time = pd.read_csv('/content/messages.csv')
date = pd.read_csv('/content/orders.csv')
```

```python
date.sample(5)
```

```python
date.info()
```

```python
#conerting to datetime datatype
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
#extract year
date['date_year'] = date['date'].dt.year
date.sample(5)
```

```python
#extract month
date['month_month_no']= date['date'].dt.month
date.sample(5)
```

```python
#extract month name
date['date_month_name'] = date['date'].dt.month_name()
date.sample(5)
```

```python
#extract days
date['date_day'] = date['date'].dt.day
date.sample()
```

```python
#day of week
date['dayofweek'] = date['date'].dt.dayofweek
date.head()
```

```python
#dayofweek- name
date['date_dayofweek'] = date['date'].dt.day_name()
date.head()
```

```python
#is weekend???
date['date_is_weekend']=np.where(date['date_dayofweek'].isin(['Sunday','Saturday']),1,0)
date.head()
```

```python
#extract week of the year

date['date_week'] = date['date'].dt.isocalendar().week
date.head()
```

```python
#extract quarter
date['date']=date['date'].dt.quarter
date.sample(5)
```
